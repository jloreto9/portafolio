```python
import pandas as pd

train_df = pd.read_csv('/kaggle/input/playground-series-s5e6/train.csv')
test_df = pd.read_csv('/kaggle/input/playground-series-s5e6/test.csv')
submission_df = pd.read_csv('/kaggle/input/playground-series-s5e6/sample_submission.csv')

print("Train DataFrame shape:", train_df.shape)
print("Test DataFrame shape:", test_df.shape)
print("Submission DataFrame shape:", submission_df.shape)

display(train_df.head())
display(test_df.head())
display(submission_df.head())
```

```python
print("Train DataFrame Info:")
train_df.info()
print("\nTest DataFrame Info:")
test_df.info()

print("\nMissing values in Train DataFrame:")
print(train_df.isnull().sum())

print("\nMissing values in Test DataFrame:")
print(test_df.isnull().sum())

print("\nDistribution of 'Fertilizer Name' in Train DataFrame:")
print(train_df['Fertilizer Name'].value_counts())

print("\nDescriptive statistics for numerical columns in Train DataFrame:")
display(train_df.describe())

print("\nDescriptive statistics for numerical columns in Test DataFrame:")
display(test_df.describe())
```

```python
# Identify categorical columns
categorical_cols = ['Soil Type', 'Crop Type']

# Apply One-Hot Encoding to training data
train_df_encoded = pd.get_dummies(train_df, columns=categorical_cols, drop_first=False)

# Apply One-Hot Encoding to testing data
test_df_encoded = pd.get_dummies(test_df, columns=categorical_cols, drop_first=False)

# Separate target variable from training features
X_train = train_df_encoded.drop('Fertilizer Name', axis=1)
y_train = train_df_encoded['Fertilizer Name']

# Align columns - this is important to ensure both dataframes have the same columns
# after one-hot encoding, handling potential differences in categories
X_train, test_df_encoded = X_train.align(test_df_encoded, join='inner', axis=1)

print("Shape of X_train after encoding and aligning:", X_train.shape)
print("Shape of test_df_encoded after encoding and aligning:", test_df_encoded.shape)
print("\nFirst 5 rows of X_train:")
display(X_train.head())
print("\nFirst 5 rows of y_train:")
display(y_train.head())
print("\nFirst 5 rows of test_df_encoded:")
display(test_df_encoded.head())
```

```python
from sklearn.model_selection import train_test_split

X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train, y_train, test_size=0.2, random_state=21, stratify=y_train
)

print("Shape of X_train_split:", X_train_split.shape)
print("Shape of X_val_split:", X_val_split.shape)
print("Shape of y_train_split:", y_train_split.shape)
print("Shape of y_val_split:", y_val_split.shape)
```

```python
from lightgbm import LGBMClassifier

# Instantiate an LGBMClassifier model
# Using default parameters to start
lgbm_model = LGBMClassifier(random_state=21)

# Train the model
lgbm_model.fit(X_train_split, y_train_split)
```

```python
import numpy as np

def apk(actual, predicted, k=3):
    """
    Calculates the Average Precision at k for a single sample.

    Args:
        actual: The ground truth label (string).
        predicted: A list of predicted labels (strings), ordered by confidence.
        k: The maximum number of predictions to consider.

    Returns:
        The Average Precision at k.
    """
    if not actual:
        return 0.0

    # Ensure actual is a list for consistency with the mapk function structure
    actual = [actual]

    score = 0.0
    num_hits = 0.0

    for i, p in enumerate(predicted[:k]):
        # Check if prediction is relevant and not a duplicate
        if p in actual and p not in predicted[:i]:
            num_hits += 1.0
            score += num_hits / (i + 1.0)

    return score / min(len(actual), k)


def mapk(actual, predicted, k=3):
    """
    Calculates the Mean Average Precision at k.

    Args:
        actual: A list of ground truth labels (strings), one per sample.
        predicted: A list of lists of predicted labels (strings), one list per sample,
                   ordered by confidence.
        k: The maximum number of predictions to consider.

    Returns:
        The Mean Average Precision at k.
    """
    return np.mean([apk(a, p, k) for a, p in zip(actual, predicted)])

print("mapk function defined.")
```

```python
# Get the predicted probabilities for each class on the validation set
y_pred_proba = lgbm_model.predict_proba(X_val_split)

# Get the list of class labels from the model
class_labels = lgbm_model.classes_

# For each sample, get the top 3 predicted class labels based on probabilities
predicted_labels = []
for probs in y_pred_proba:
    # Get the indices that would sort the probabilities in descending order
    top_k_indices = np.argsort(probs)[::-1][:3]
    # Get the corresponding class labels
    top_k_labels = [class_labels[i] for i in top_k_indices]
    predicted_labels.append(top_k_labels)

# Convert the actual validation labels to a list
actual_labels = y_val_split.tolist()

# Calculate the MAP@3 score
mapk_score = mapk(actual_labels, predicted_labels, k=3)

# Print the calculated MAP@3 score
print(f"Mean Average Precision @ 3 (MAP@3) on the validation set: {mapk_score:.4f}")
```
